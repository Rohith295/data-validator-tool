# Sample Files Walkthrough

Validate a few files from `sample_data/` using `validate_file()`.

This is a practical notebook for checking how the library behaves against the same sample inputs used by the CLI.

```python
from pathlib import Path

from data_validator import ReportView, validate_file
```

```python
ROOT = Path.cwd()
if not (ROOT / "sample_data").exists():
    ROOT = ROOT.parent

ROOT
```

```python
good_csv = ROOT / "sample_data" / "1.csv"
bad_csv = ROOT / "sample_data" / "4.csv"
default_schema = ROOT / "sample_data" / "my_schema.json"
messy_csv = ROOT / "sample_data" / "messy.csv"
messy_schema = ROOT / "sample_data" / "messy_schema.json"

good_csv, bad_csv, default_schema
```

```python
good_report = validate_file(good_csv, default_schema)
bad_report = validate_file(bad_csv, default_schema)
messy_report = validate_file(messy_csv, messy_schema)
```
